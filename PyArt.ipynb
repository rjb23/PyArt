```python
import numpy as np
import matplotlib.pyplot as plt

# Define image size
height = 400
width = 400

# Create an empty image array
img = np.zeros((height, width), dtype=np.uint8)

# Draw face features using grayscale pixel values
# Draw the outline of the face
for x in range(100, 300):
    img[100, x] = 255
    img[299, x] = 255
    img[x, 100] = 255
    img[x, 299] = 255

# Draw the eyes
for x in range(150, 170):
    img[150, x] = 255
    img[150, x+40] = 255

# Draw the nose
for y in range(175, 210):
    img[y, 200] = 255

# Draw the mouth
for x in range(150, 170):
    img[230, x] = 255

# Display the image
plt.imshow(img, cmap='gray')
plt.show()
```

```python
import random
import numpy as np
import cv2

# Create a black image
img = np.zeros((500, 500, 3), np.uint8)

# Draw 100 circles with random colors and positions
for i in range(100):
    # Randomly generate a color
    color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))
    # Randomly generate a position and radius
    center = (random.randint(0, 499), random.randint(0, 499))
    radius = random.randint(5, 100)
    # Draw the circle
    cv2.circle(img, center, radius, color, -1)

# Display the image
cv2.imshow('Random Circles', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from PIL import Image, ImageDraw

# Create a new image with a white background
img = Image.new('RGB', (400, 400), color='white')

# Draw the monkey
draw = ImageDraw.Draw(img)
draw.rectangle((50, 50, 350, 350), fill=(200, 200, 200), outline='black')
draw.ellipse((150, 100, 250, 200), fill='brown', outline='black')
draw.ellipse((150, 250, 250, 350), fill='brown', outline='black')
draw.ellipse((100, 150, 150, 200), fill='brown', outline='black')
draw.ellipse((250, 150, 300, 200), fill='brown', outline='black')
draw.ellipse((200, 225, 250, 275), fill='black', outline='black')
draw.line((200, 225, 250, 225), fill='black', width=2)

# Save the image
img.save('monkey.png')
```

```python
from PIL import Image, ImageDraw

# Create a new image with a black background
img = Image.new("RGB", (400, 400), (0, 0, 0))

# Create a drawing context
draw = ImageDraw.Draw(img)

# Draw the monkey's head
draw.ellipse((100, 50, 300, 250), fill=(186, 157, 118), outline=(0, 0, 0))

# Draw the monkey's eyes
draw.ellipse((140, 90, 160, 110), fill=(255, 255, 255), outline=(0, 0, 0))
draw.ellipse((240, 90, 260, 110), fill=(255, 255, 255), outline=(0, 0, 0))
draw.ellipse((143, 93, 157, 107), fill=(0, 0, 0), outline=(0, 0, 0))
draw.ellipse((243, 93, 257, 107), fill=(0, 0, 0), outline=(0, 0, 0))

# Draw the monkey's nose
draw.polygon([(200, 150), (195, 170), (205, 170)], fill=(120, 70, 0), outline=(0, 0, 0))

# Draw the monkey's mouth
draw.arc((140, 170, 260, 220), start=0, end=180, fill=(255, 255, 255), width=5)

# Save the image
img.save("monkey.png")

# Display the image
img.show()
```

```python
import numpy as np
import cv2

# Create a black image
img = np.zeros((512, 512, 3), np.uint8)

# Draw a circle in the middle of the image
cv2.circle(img, (256, 256), 128, (0, 255, 0), -1)

# Draw two eyes
cv2.circle(img, (200, 200), 30, (255, 255, 255), -1)
cv2.circle(img, (312, 200), 30, (255, 255, 255), -1)

# Draw a nose
cv2.rectangle(img, (225, 280), (287, 320), (0, 0, 255), -1)

# Draw a mouth
cv2.rectangle(img, (225, 350), (287, 380), (255, 0, 0), -1)

# Display the image
cv2.imshow("Ape", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Create a black image
img = np.zeros((500, 500, 3), np.uint8)

# Draw a green rectangle for the forest
cv2.rectangle(img, (0, 0), (500, 350), (0, 255, 0), -1)

# Draw a blue sky
cv2.rectangle(img, (0, 0), (500, 200), (255, 0, 0), -1)

# Draw a white circle for the sun
cv2.circle(img, (100, 100), 50, (255, 255, 255), -1)

# Draw a running boy with yellow clothes and black hair
cv2.circle(img, (300, 400), 50, (0, 255, 255), -1)
cv2.rectangle(img, (275, 350), (325, 400), (0, 0, 0), -1)
cv2.line(img, (325, 375), (350, 350), (0, 0, 0), 5)

# Show the image
cv2.imshow("Image", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
